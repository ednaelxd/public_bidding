### Feature Engeneering

`Ajuste dos numéricos`

```python
df.Total=df.Total.str.replace(',','.')
df['Total'] = df['Total'].astype(float)

df.valor_item = df.valor_item.str.replace(',','.')
df['valor_item'] = df['valor_item'].astype(float)
```

`Nova Feature - Valor Individual dos itens`

```python
df['valor_individual'] = df.valor_item / df['Quantidade Item']
```

`Tentativa de agrupamento mensal`

```python
df.data = pd.to_datetime(df.data)
df2 = df.resample(rule='M', on='data').sum()
```

```python
df.head(1)
```

`Agrupamento por parâmetros estatísticos`

```python
agrup = df.groupby(['Descrição','UF','data']).agg({'valor_individual':['count','min','max','mean','std']})
```

```python
agrup = pd.DataFrame(agrup['valor_individual']).reset_index(drop=False)
```

```python
agrup = agrup.sort_values(by='count',ascending=False)
agrup
```

## Investigação de Anomalias - Abordagem estatística
### Parâmetro - Descrição [base de itens]
- Essa abordagem, talvez seja mais sucetivel ao agrupamento de itens diferentes numa mesma categoria, por falta de detalhamento na descrição
- Recorte para remoção dos serviços, por não caracterizarem comparação de iguais

```python

agrup = df[df['Obras']=='Compras/Servicos'].groupby(['Descrição','UF']).agg({'valor_individual':['count','min','max','mean','std']})
agrup = pd.DataFrame(agrup['valor_individual']).reset_index(drop=False)
agrup = agrup.sort_values(by='count',ascending=False)
agrup
```

`Recorte dos dados que possuem desvio padrão válido`
- Não nulos com número de compras superior a 10

```python
range1 = agrup[(agrup['mean']!=0) & (agrup['count']>=10)]
range1.shape
```

```python
range1 = range1.sort_values(by='count',ascending=False)
```

```python
pd.options.display.max_colwidth = 500
pd.set_option('display.float_format', lambda x: '%.5f' % x)
range1[0:20]
```

```python
# Avaliando por período

fig = px.box(range1,x='data',y='std',hover_name='Descrição',color='UF')
fig.show()
```

Alguns itens chegam a registrar desvio padrão de 600x

```python
# Avaliando por item


fig = px.box(range1,x='Descrição',y='mean',hover_name='Descrição',color='UF')
fig.show()
```

`Panorama geral da discrepância de valores dos itens`

```python
import plotly.express as px

fig = px.treemap(range1,path=[px.Constant('ASADasdasd'),'Descrição','UF'],
                 values='std',maxdepth=3)
 #           values='Total de Licitações')
fig.update_layout(uniformtext_minsize=18)
fig.update_traces(root_color="lightgray",textfont_color='rgb(255,255,255)')
```

```python
# Desconsiderando os estados

agrup = df[df['Obras']=='Compras/Servicos'].groupby(['Descrição','data']).agg({'valor_individual':['count','min','max','mean','std']})
agrup = pd.DataFrame(agrup['valor_individual']).reset_index(drop=False)
agrup = agrup.sort_values(by='count',ascending=False)

range1 = agrup[(agrup['mean']!=0) & (agrup['count']>=10)]
range1.shape
range1 = range1.sort_values(by='std',ascending=False)
```

```python
# Avaliando por período

fig = px.box(range1,x='Descrição',y='mean',hover_name='count')   #,color='Descrição')
fig.show()
```
